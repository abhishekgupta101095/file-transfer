```python
# Import the required libraries
from sentence_transformers import SentenceTransformer, util
import torch

import pandas as pd
```

```python
# Load the MPNet model
model = SentenceTransformer('sentence-transformers/all-mpnet-base-v2')
```

```python
df = pd.read_excel(r"C:\Users\abhishek.cw.gupta\Desktop\UserQueryExample.xlsx")
df
```

```python
df = df[['User Query', 'Solution']]
df
```

```python
corpus = df['User Query'].tolist()
```

```python
corpus
```

```python
# Encode the corpus with the model
corpus_embeddings = model.encode(corpus)
```

```python
mylist = []
for items in corpus_embeddings:
    mylist.append(items)
```

```python
df['Embeddings'] = mylist
```
